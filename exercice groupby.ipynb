```python
import pandas as pd
import numpy as np

df_user_usage = pd.read_csv("user_usage.csv")
df_user_device = pd.read_csv("user_device.csv")
df_android = pd.read_csv("android_devices.csv")
```

```python
df_android.rename(columns={"Retail Branding": "Branding"}, inplace=True)
```

```python
df_first_merge = pd.merge(df_user_usage,df_user_device[['use_id','platform','device']],on='use_id', how='inner')
df_first_merge
```

```python
print(df_first_merge.shape)
print("-----")
print(df_user_device.shape)
print(df_user_usage.shape)

print("on remarque que le merge inner ne récupère que les colonnes ou use_id est le même à droite et à gauche")
```

```python
print(df_first_merge.value_counts("use_id"))
print("")
print("les use_id non présent après le merge : ")
df_non_match = pd.merge(df_user_usage,df_user_device[['use_id','platform','device']], how="outer", on='use_id', indicator=True)
df_non_match = df_non_match[(df_non_match._merge != 'both')]
print(df_non_match.value_counts("use_id"))
```

```python
df_left_merge = pd.merge(df_user_usage,df_user_device[['use_id','platform','device']], how="left", on='use_id')
print(df_left_merge.shape)
print("-----")
print(df_user_usage.shape)
print("-----")
print("valeurs manquantes : ")
print(df_left_merge.device.isna().sum())
print("-----")
df_left_merge
```

```python
# right merge

df_right_merge = pd.merge(df_user_usage,df_user_device[['use_id','platform','device']], how="right", on='use_id')

print("user_device dimensions : {}".format(df_user_device.shape))
print("result dimensions : {}".format(df_right_merge.shape ))
print("There are {} missing values in the 'monthly_mb' column in the result.".format(df_right_merge.monthly_mb.isna().sum()))
print("There are {} missing values in the 'platform' column in the result.".format(df_right_merge.platform.isna().sum()))
```

```python
a = df_user_usage["use_id"].unique()
b = df_user_device["use_id"].unique()
c = np.concatenate((a, b))

df_outer_merge = pd.merge(df_user_usage,df_user_device[['use_id','platform','device']], how="outer", on='use_id')

print("There are {} unique values of use_id in our dataframes.".format(len(np.unique(c))))
print("Outer merge result has {} rows.".format(df_outer_merge.shape))
print("There are {} rows with no missing values.".format(len(np.unique(c)) - (df_outer_merge.shape[0] - df_outer_merge.dropna().shape[0])))
```
